```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Banking.csv')
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
#Generate descriptive statistics for the dataframe
df.describe()
```

```python
# Estimated Income

bins = [0,100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']
df ['Income Band']= pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
df.head()
```

```python
df['Income Band'].value_counts().plot(kind="bar")
```

##**Distribution**

```python
#Examine the distribution of unique categories in categorical columns
categorical_cols = df[['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Properties Owned', 'Loyalty Classification', 'Risk Weighting', 'Income Band']].columns
for col in categorical_cols:
    print(f"Unique categories in {col}:")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df):
  plt.figure(i)
  sns.countplot(data=df, x= predictor, hue = 'GenderId')
```

```python
for i, predictor in enumerate( df[['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Properties Owned', 'Loyalty Classification', 'Risk Weighting', 'Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df, x= predictor, hue = 'Nationality')
```

##**Histplot**

```python
#HIstplot of value counts for different Occupation

for col in categorical_cols:
  if col == 'Occupation':
      continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

##**Numerical Analysis**

```python
numerical_cols = ["Estimated Income", "Superannuation Savings", "Credit Card Balance", "Bank Loans", "Bank Deposits", "Checking Accounts", "Saving Accounts", "Foreign Currency Account", "Business Lending"]

#University analysis and visualization
plt.figure(figsize=(6, 20))
for i,col in enumerate(numerical_cols):
  plt.subplot(9,1,i+1)
  sns.histplot(df[col],kde= True)
  plt.title(col)
plt.tight_layout() # Added to prevent plot overlap
plt.show()
```

##**Heatmaps**

```python
numerical_cols = ["Estimated Income", "Superannuation Savings", "Credit Card Balance", "Bank Loans", "Bank Deposits", "Checking Accounts", "Saving Accounts", "Foreign Currency Account", "Business Lending"]

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="crest", fmt = ".2f")
plt.title("Correlation Heatmap")
plt.show()
```

##**Insights of EDA**

1. The strongest correlation occur between bank deposits with "Checking Accounts", "Saving Accounts", "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts
